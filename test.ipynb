```python
with open("newpose.txt", 'r') as f:
    data = f.readlines()
    data = [[word for word in line.strip().split(' ')[1:]] for line in data]
    data = [line  for i,line in enumerate(data) if i % 20 == 0]

data = data[: int(len(data) / 2)][:20]
with open("pose.txt", 'w') as f:
    for line in data:
        newline = line[3:] + line[:3]
        ss = ' '.join(newline) + "\n"
        f.write(ss)
```
